#### e. Cantidad de registros duplicados exactos.



```python
resumen_duplicados = pd.DataFrame({
    'Métrica': [
        'Duplicados exactos (todas las ocurrencias, keep=False)',
        'Duplicados exactos (sin la primera ocurrencia)',
        'Filas completamente vacías (todas las columnas nulas)',
        'Duplicados exactos entre registros CON datos',
        'Códigos (CODIGO) repetidos entre registros no nulos',
    ],
    'Cantidad': [
        df.duplicated(keep=False).sum(),
        df.duplicated().sum(),
        df.isnull().all(axis=1).sum(),
        df[~df.isnull().all(axis=1)].duplicated().sum(),
        df[df['CODIGO'].notna()].duplicated(subset=['CODIGO']).sum(),
    ]
})
resumen_duplicados
```

`duplicated()` marca 43 registros como duplicados exactos (44 si se cuentan todas las ocurrencias con `keep=False`). Sin embargo, al revisarlos se determinó que **la totalidad corresponde a filas completamente vacías** (sin ningún valor en sus 17 columnas): son las mismas ~44 filas responsables de los valores faltantes reportados en el inciso c, lo cual explica por qué casi todas las variables comparten un 0.19% de nulos.

Al excluir esas filas vacías, no queda ningún duplicado exacto entre establecimientos con información real, y la variable CODIGO —que actúa como identificador— no presenta ningún valor repetido entre los registros no nulos. Este hallazgo también implica que las filas vacías deberán eliminarse (o al menos documentarse) en el plan de limpieza, ya que están inflando artificialmente tanto el conteo de faltantes como el de duplicados.

#### f. Variables con valores fuera de dominio o inconsistentes.



```python
from IPython.display import display

def tabla_conteo(serie, nombre_col):
    return serie.value_counts(dropna=False).rename_axis(nombre_col).reset_index(name='Cantidad')

print("AREA: valores fuera del dominio esperado {URBANA, RURAL}")
display(tabla_conteo(df['AREA'], 'AREA'))

print("NIVEL: se esperaba solo DIVERSIFICADO según el alcance del proyecto")
display(tabla_conteo(df['NIVEL'], 'NIVEL'))

print("STATUS: categorías y su frecuencia")
display(tabla_conteo(df['STATUS'], 'STATUS'))

telefonos = df['TELEFONO'].dropna().astype(str)
solo_digitos = telefonos.str.fullmatch(r"\d+")
telefono_valido = solo_digitos & (telefonos.str.len() == 8)
patron_codigo = df['CODIGO'].dropna().str.fullmatch(r"\d{2}-\d{2}-\d{4}-\d{2}")

print("TELEFONO y CODIGO: cumplimiento de dominio esperado")
resumen_dominio = pd.DataFrame({
    'Variable': ['TELEFONO', 'TELEFONO', 'CODIGO'],
    'Regla de dominio': [
        'Solo caracteres numéricos',
        'Exactamente 8 dígitos (formato guatemalteco)',
        'Patrón NN-NN-NNNN-NN',
    ],
    'Cumplen': [solo_digitos.sum(), telefono_valido.sum(), patron_codigo.sum()],
    'NO cumplen': [(~solo_digitos).sum(), len(telefonos) - telefono_valido.sum(), (~patron_codigo).sum()],
    'Total evaluado': [len(telefonos), len(telefonos), len(patron_codigo)],
})
display(resumen_dominio)

print("TELEFONO: distribución de longitud entre los valores puramente numéricos")
tabla_longitud = telefonos[solo_digitos].str.len().value_counts().rename_axis('N° de dígitos').reset_index(name='Cantidad').sort_values('N° de dígitos')
display(tabla_longitud)
```

AREA incluye la categoría **"SIN ESPECIFICAR"** (2 registros), fuera del dominio binario esperado {URBANA, RURAL}. NIVEL contiene **BASICO** (13,552 registros) además de DIVERSIFICADO (9,706), lo cual está fuera del alcance declarado por el proyecto (solo centros que llegan a diversificado); esto sugiere que la extracción de datos no filtró exclusivamente por ese nivel y deberá decidirse en el plan de limpieza si se filtra o se documenta como parte del universo de datos. STATUS presenta dos categorías poco frecuentes (TEMPORAL NOMBRAMIENTO y TEMPORAL TITULOS, con 110 y 44 registros) que son válidas pero merecen revisión por su baja representación.

La variable con más problemas de dominio es **TELEFONO**: 249 registros contienen caracteres no numéricos (guiones, comas o texto como "CEL."), y entre los valores puramente numéricos hay longitudes de 1, 2, 4, 5, 6, 7, 9, 10 y hasta 16 dígitos en lugar de los 8 dígitos esperados para un teléfono guatemalteco.

En contraste, DEPARTAMENTO (validado en el inciso d) y CODIGO sí cumplen su dominio esperado: los 22 valores de DEPARTAMENTO coinciden exactamente con los 22 departamentos oficiales de Guatemala, y el 100% de los códigos no nulos siguen el patrón NN-NN-NNNN-NN.

#### g. Variables con formatos inconsistentes.



```python
from IPython.display import display

print("TELEFONO: ejemplos de formatos mixtos (guiones, comas, texto embebido)")
no_numericos = telefonos[~solo_digitos]
tabla_telefono_ejemplos = no_numericos.head(10).reset_index(drop=True).rename('Valor original').to_frame()
display(tabla_telefono_ejemplos)

est = df['ESTABLECIMIENTO'].dropna().astype(str)
print("ESTABLECIMIENTO: estilo de comillas usado para encerrar apodos")
tabla_comillas = pd.DataFrame({
    'Estilo de comillas': ['Dobles ("...")', "Simples ('...')"],
    'Cantidad': [est.str.contains(chr(34)).sum(), est.str.contains(chr(39)).sum()],
})
display(tabla_comillas)

dire = df['DIRECCION'].dropna().astype(str)
con_minusculas = dire.str.contains('[a-z]', regex=True)
print("DIRECCION: consistencia de mayúsculas")
tabla_direccion = pd.DataFrame({
    'Métrica': ['Con alguna letra minúscula', 'Total de direcciones no nulas'],
    'Cantidad': [con_minusculas.sum(), len(dire)],
})
display(tabla_direccion)

print("DIRECCION: ejemplos con minúsculas (abreviaturas de ordinal inconsistentes)")
tabla_direccion_ejemplos = dire[con_minusculas].head(5).reset_index(drop=True).rename('Valor original').to_frame()
display(tabla_direccion_ejemplos)
```

Además de los problemas de dominio, TELEFONO mezcla distintos formatos dentro de la misma columna: números separados por guion (p. ej. "66420121-66420111"), por coma (p. ej. "6372043, 6372123") o acompañados de texto (p. ej. "6400403 CEL. 4019799"), lo que indica que el campo a veces almacena más de un contacto o un número con anotación en lugar de un solo teléfono.

ESTABLECIMIENTO no tiene un estilo uniforme para encerrar apodos o nombres comerciales: 3,696 registros usan comillas dobles y 1,526 usan comillas simples. DIRECCION es casi enteramente uniforme en mayúsculas (solo 19 de 22,988 registros contienen alguna letra minúscula), pero esos mismos registros muestran una abreviatura inconsistente del ordinal de calle/avenida (p. ej. "6a. CALLE" en minúscula frente a la convención dominante "6A. CALLE"), evidenciando que la normalización de texto (inciso 5c del proyecto) tendrá que estandarizar mayúsculas y el uso de puntos en las abreviaturas.

#### h. Identificación de problemas potenciales de calidad de datos.



```python
import unicodedata
from IPython.display import display

def normalizar(texto):
    return ''.join(c for c in unicodedata.normalize('NFD', str(texto)) if unicodedata.category(c) != 'Mn').upper().strip()

# DEPARTAMENTAL subdivide GUATEMALA en 4 zonas y QUICHE en 2; validamos que
# siempre correspondan al DEPARTAMENTO correcto (consistencia entre variables)
sub = df[['DEPARTAMENTO', 'DEPARTAMENTAL']].dropna().copy()
sub['dep_norm'] = sub['DEPARTAMENTO'].map(normalizar)
sub['depal_norm'] = sub['DEPARTAMENTAL'].map(normalizar)
sub['coincide'] = sub.apply(lambda r: r['depal_norm'].startswith(r['dep_norm']), axis=1)

print("Consistencia cruzada: DEPARTAMENTO vs. DEPARTAMENTAL")
tabla_consistencia = pd.DataFrame({
    'Métrica': ['Registros comparados', 'Contradicciones encontradas'],
    'Cantidad': [len(sub), (~sub['coincide']).sum()],
})
display(tabla_consistencia)

print("Síntesis de problemas potenciales de calidad de datos")
resumen_problemas = pd.DataFrame([
    {'Problema': 'Filas completamente vacías', 'Variable(s)': 'Todas', 'Cantidad': 44,
     'Detalle': 'Inflan los conteos de faltantes (c) y de duplicados exactos (e)'},
    {'Problema': 'Tipo de dato incorrecto', 'Variable(s)': 'TELEFONO, CODIGO', 'Cantidad': 23302,
     'Detalle': 'Todas las columnas se leyeron como texto (b)'},
    {'Problema': 'Formato/longitud inválida', 'Variable(s)': 'TELEFONO', 'Cantidad': 249,
     'Detalle': 'Guiones, comas, texto embebido; longitudes de 1 a 28 en vez de 8 dígitos'},
    {'Problema': 'Fuera del alcance declarado', 'Variable(s)': 'NIVEL', 'Cantidad': 13552,
     'Detalle': 'Incluye BASICO; el proyecto pide solo nivel DIVERSIFICADO'},
    {'Problema': 'Categoría residual fuera de dominio', 'Variable(s)': 'AREA', 'Cantidad': 2,
     'Detalle': '"SIN ESPECIFICAR" no pertenece a {URBANA, RURAL}'},
    {'Problema': 'Formato de texto no uniforme', 'Variable(s)': 'ESTABLECIMIENTO, DIRECCION', 'Cantidad': 5241,
     'Detalle': 'Comillas simples/dobles mezcladas; mayúsc./minúsc. en abreviaturas de calle'},
    {'Problema': 'Consistencia entre variables', 'Variable(s)': 'DEPARTAMENTO / DEPARTAMENTAL', 'Cantidad': 0,
     'Detalle': 'Sin contradicciones — hallazgo positivo, no requiere corrección'},
])
display(resumen_problemas)
```

Al cruzar DEPARTAMENTO con DEPARTAMENTAL no se encontró ninguna contradicción: las 26 categorías de DEPARTAMENTAL siempre corresponden al departamento correcto en DEPARTAMENTO. Es una validación cruzada exitosa entre dos variables relacionadas, y descarta que la granularidad extra de DEPARTAMENTAL (visible en el inciso d como "26 valores únicos" frente a los 22 departamentos) sea un error de inconsistencia; es, en realidad, una subdivisión administrativa válida (Guatemala en 4 zonas, Quiché en 2).

En síntesis, los principales problemas de calidad identificados en este diagnóstico son:

1. **Filas completamente vacías** (44 registros, ~0.19%) que inflan artificialmente los conteos de valores faltantes (inciso c) y de duplicados exactos (inciso e).
2. **Tipos de dato incorrectos**: todas las columnas fueron leídas como texto (inciso b); TELEFONO debería ser numérico y CODIGO podría descomponerse en variables categóricas (departamento-distrito-consecutivo-nivel).
3. **TELEFONO es el campo más problemático**: mezcla números concatenados, separadores inconsistentes (guion, coma, espacio), texto embebido ("CEL.") y longitudes que van de 1 a 28 caracteres en lugar de los 8 dígitos esperados.
4. **NIVEL incluye BASICO** pese a que el alcance del proyecto solo pide establecimientos de nivel diversificado — deberá decidirse si se filtra o se documenta como parte del universo de datos.
5. **Categoría residual "SIN ESPECIFICAR"** en AREA, fuera del dominio binario esperado (URBANA/RURAL).
6. **Formato de texto no uniforme** en ESTABLECIMIENTO (comillas simples vs. dobles) y en DIRECCION (abreviaturas de ordinal en mayúscula/minúscula), aunque en general el dataset ya está mayormente normalizado a mayúsculas.
7. **Ausencia de contradicciones** entre DEPARTAMENTO y DEPARTAMENTAL, y entre los 22 valores de DEPARTAMENTO y la división política oficial del país — puntos positivos que no requieren corrección.

Estos hallazgos alimentarán directamente el plan de limpieza (actividad 4), donde deberá definirse la regla de corrección para cada uno de estos problemas.
