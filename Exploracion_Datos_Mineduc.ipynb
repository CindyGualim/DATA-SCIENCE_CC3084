#### g. Variables con formatos inconsistentes.



```python
est = df['ESTABLECIMIENTO'].dropna().astype(str)
dire = df['DIRECCION'].dropna().astype(str)
distrito = df['DISTRITO'].dropna().astype(str)
formato_distrito = distrito.str.fullmatch(r'\d{2}-\d{3}') | distrito.str.fullmatch(r'\d{2}-\d{2}-\d{4}')

pd.DataFrame({
    'Variable': ['ESTABLECIMIENTO', 'ESTABLECIMIENTO', 'DIRECCION', 'TELEFONO', 'DISTRITO'],
    'Problema de formato': [
        'Usa comillas dobles', 'Usa comillas simples', 'Contiene minúsculas',
        'Contiene guion, coma o texto', 'No sigue NN-NNN ni NN-NN-NNNN',
    ],
    'Cantidad': [
        est.str.contains(chr(34)).sum(),
        est.str.contains(chr(39)).sum(),
        dire.str.contains('[a-z]', regex=True).sum(),
        (~telefonos.str.fullmatch(r'\d+')).sum(),
        (~formato_distrito).sum(),
    ]
})
```

Además de los problemas de dominio, TELEFONO mezcla distintos formatos dentro de la misma columna: números separados por guión, por coma o acompañados de texto, lo que indica que el campo a veces almacena más de un contacto en lugar de un solo teléfono.

ESTABLECIMIENTO no tiene un estilo uniforme para encerrar apodos o nombres comerciales: 2,028 registros usan comillas dobles y 508 usan comillas simples. DIRECCION es casi enteramente uniforme en mayúsculas (solo 7 de 9,637 registros contienen alguna letra minúscula), pero esos mismos registros muestran una abreviatura inconsistente del ordinal de calle/avenida (p. ej. "8a. AVE." en minúscula frente a la convención dominante "8A. AVE."), evidenciando que la normalización de texto tendrá que estandarizar mayúsculas y el uso de puntos en las abreviaturas.

DISTRITO también mezcla dos formatos estructurales distintos: 3,928 registros siguen el patrón NN-NNN (p. ej. "01-406") y 5,461 siguen NN-NN-NNNN (p. ej. "01-01-0146"), además de 16 valores truncados sin dígitos después del guion (p. ej. "01-"). Esto sugiere que los datos provienen de al menos dos convenciones de codificación distintas y deberá decidirse a cuál estandarizar antes de usar esta variable.

#### h. Identificación de problemas potenciales de calidad de datos.



```python
import unicodedata
normalizar = lambda s: ''.join(c for c in unicodedata.normalize('NFD', str(s)) if unicodedata.category(c) != 'Mn').upper()

# DEPARTAMENTAL subdivide GUATEMALA (4 zonas) y QUICHE (2); validamos que
# siempre correspondan al DEPARTAMENTO correcto (consistencia entre variables)
sub = df[['DEPARTAMENTO', 'DEPARTAMENTAL']].dropna()
dep_norm, depal_norm = sub['DEPARTAMENTO'].map(normalizar), sub['DEPARTAMENTAL'].map(normalizar)
coincide = [d.startswith(dep) for d, dep in zip(depal_norm, dep_norm)]

pd.DataFrame({
    'Métrica': ['Registros comparados', 'Contradicciones'],
    'Cantidad': [len(sub), (~pd.Series(coincide)).sum()]
})
```

Al cruzar DEPARTAMENTO con DEPARTAMENTAL no se encontró ninguna contradicción: las 26 categorías de DEPARTAMENTAL siempre corresponden al departamento correcto en DEPARTAMENTO. Es una validación cruzada exitosa entre dos variables relacionadas, y descarta que la granularidad extra de DEPARTAMENTAL (visible en el inciso d como "26 valores únicos" frente a los 22 departamentos) sea un error de inconsistencia; es, en realidad, una subdivisión administrativa válida (Guatemala en 4 zonas, Quiché en 2).

## 4. Plan de limpieza

Antes de modificar los datos, se define a continuación un plan de limpieza para cada variable, basado en los hallazgos del diagnóstico (inciso 3). Como paso previo y transversal a todas las variables, se eliminarán las **22 filas completamente vacías** (inciso 3e). Las cantidades de "problemas encontrados" de la tabla ya excluyen esas 22 filas cuando corresponde, para no contar el mismo problema dos veces.

| Variable | a. Problemas encontrados | b. Regla de corrección | c. Riesgos |
